```python
######################ğ¶ğ‘œğ‘›ğ‘ ğ‘¡ğ‘Ÿğ‘ğ‘ ğ‘¡###############################
def contrast(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s=s+x[i,j]*(i-j)**2
    return(s)        
######################ğ¶ğ‘œğ‘Ÿğ‘Ÿğ‘’ğ‘™ğ‘ğ‘¡ğ‘–ğ‘œğ‘›##################
def correlation(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s=s+x[i,j]*(((i-np.mean(x[i,:]))*(j-np.mean(x[:,j])))/((np.var(x[i,:]))*(np.var(x[:,j]))))
    return(s)  
#################################ğ‘ƒğ‘Ÿğ‘œğ‘šğ‘–ğ‘›ğ‘’ğ‘›ğ‘ğ‘’ ğ¶ğ‘™ğ‘¢ğ‘ ğ‘¡ğ‘’ğ‘Ÿğ‘–ğ‘›ğ‘”################
def pro_cluster(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s=s+(((i-np.mean(x[i,:]))**4*(j-np.mean(x[:,j]))**4)/((np.var(x[i,:]))**2*(np.var(x[:,j]))**2))
    return(s)  
#################################ğ‘†â„ğ‘ğ‘‘ğ‘œğ‘¤ ğ¶ğ‘™ğ‘¢ğ‘ ğ‘¡ğ‘’ğ‘Ÿğ‘–ğ‘›ğ‘”################
def sha_cluster(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s=s+(((i-np.mean(x[i,:]))**3*(j-np.mean(x[:,j]))**3)/((np.var(x[i,:]))*(np.std(x[i,:]))*(np.var(x[:,j]))*(np.std(x[:,j]))))
    return(s)
#################################ğ·ğ‘–ğ‘ ğ‘ ğ‘–ğ‘šğ‘–ğ‘™ğ‘–ğ‘Ÿğ‘–ğ‘¡ğ‘¦################
def dissimiler(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s=s+x[i,j]*abs(i-j)
    return(s)    
#################################ğ¸ğ‘›ğ‘’ğ‘Ÿğ‘”ğ‘¦################
def energy(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s=s+x[i,j]*2
    return(math.sqrt(s))     
#################################ğ¸ğ‘›ğ‘¡ğ‘Ÿğ‘œğ‘ğ‘¦################
def entropy(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            if(x[i,j]>0):
                s=s+x[i,j]*(-math.log(x[i,j]))
                
            
    return(s)   
#################################ğ»ğ‘œğ‘šğ‘œğ‘”ğ‘’ğ‘›ğ‘’ğ‘¡ğ‘¦################
def homogenety(x):
    s=0
    for i in range(x.shape[0]):
        for j in range(x.shape[1]):
            s=s+x[i,j]/(1+(i-j)**2)
    return(s)
#################################Maximun Probability################
def max_prob(x):
    
    return(np.max(x))    
        
               
        
```

```python

##########################################NDI#############################
def NDI(gre , red):
    numerator = np.subtract(gre, red)
    denominator = np.add(gre, red)
    denominator[denominator==0]=0.01
    result = np.divide(numerator, denominator)
    result=np.divide(result-np.amin(result),np.amax(result)-np.amin(result))
    return(result)
##########################################SR#############################
def SR(red , blue):
    numerator1 = np.subtract(red, blue)
    denominator1 = np.add(red, blue)
    denominator1[denominator1==0]=0.01
    result1 = np.divide(numerator1, denominator1)
    numerator = np.add(1, result1)
    denominator = np.subtract(1, result1)
    denominator[denominator==0]=0.01
    result2 = np.divide(numerator, denominator)
    kernel = np.ones((7,7), np.uint8)
    # Erosion and dilation
    erode = cv2.erode(result2,kernel,iterations = 1)
    dilate = cv2.dilate(erode, kernel,iterations = 2)
    result = cv2.GaussianBlur(dilate,(3,3),1)
    result=np.divide(result-np.amin(result),np.amax(result)-np.amin(result))
    return(result)
##########################################VARI#############################

    
```
